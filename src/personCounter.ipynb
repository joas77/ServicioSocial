# Contador de personas

Este proyecto es un contador de personas utilizando openCV.
El proyecto consiste de las siguientes etapas:
* Obtención de imagenes 
* Filtrado de imagenes
* Detección de regiones
* Procesamiento de las regiones

## Importando paquetes

```python
#importing some useful packages#import 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
%matplotlib inline
```

## Funciones auxiliares

```python
def show_image(image, title, cmap=None):
    plt.figure()
    plt.imshow(image, cmap)
    plt.title(title)
```

## Leer una imagen

```python
image = mpimg.imread("../imagenes/blob.jpg")
show_image(image, "blob original")

gray_im = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
show_image(gray_im, "blob escala de grises", cmap="gray")
```

```python
detector = cv2.SimpleBlobDetector_create()

# Detect blobs.
keypoints = detector.detect(image)
 
# Draw detected blobs as red circles.
# cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS ensures the size of the circle corresponds to the size of blob
im_with_keypoints = cv2.drawKeypoints(image, keypoints, np.array([]), (0,0,255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
 
# Show keypoints
show_image(im_with_keypoints, "blobs")
```
